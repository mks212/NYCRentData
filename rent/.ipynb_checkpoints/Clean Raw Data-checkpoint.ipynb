```python
df = df.loc[pd.notnull(df['neighborhood'])] #keep everything except the one missing a hood
```

```python
#Get all possible amenties
amens = df['amenities'].str.split(";")
amens.dropna(inplace=True)
amens = list(chain(*amens))
#sqft = [x for x in amens if "ft" in x]
amens = [x for x in amens if "ft" not in x] #remove sq ft
set(amens)
```

```python
#Since it is a limited list, we will add special columns one at a time for Doorman, Elevator, Furnished, and Laundry
df['elevator'] = df['amenities'].str.contains('Elevator', na=False, regex=False) #regex gives performance boost
df['doorman'] = df['amenities'].str.contains('Doorman', na=False, regex=False)
df['laundry'] = df['amenities'].str.contains('Laundry', na=False, regex=False)
df['furnished'] = df['amenities'].str.contains('Furnished', na=False, regex=False )
```

```python
#used this code to get a data frame for stack overflow question
#stack = df[['address', 'amenities']]
#stack.head(4).to_dict()
```

```python
df.sample(10)
```

```python
#Get sqft info
df['amenities'] = df['amenities'].str.replace(",","") #remove commas, for example, 1,100 to 1100
df['sqft'] = df['amenities'].str.extract('(\d+)')
```

```python
df.sample(10) #make sure all amenities are done
```

```python
#Get neighborhood and borough. Neighborhood is first item in list of 3, borough is last
neighborhoods = df['neighborhood'].str.split(";")
df['borough'] = [x[-1] for x in neighborhoods]
df['borough'] = df['borough'].str.strip()

df['hood'] = [x[0] for x in neighborhoods]
df['hood'] = df['hood'].str.strip()
```

```python
df.sample(10) #make sure neighborhood and borough are done.
```

```python
df = df.drop(['amenities', 'neighborhood', 'broker'], 1)
```

```python
df.sort_values(by = ['address']) #examine duplicates
```

```python
df = df.drop_duplicates(subset=['address', 'baths', 'beds', 'rent'])
```

```python
df = df.reindex(columns = ['address', 'rent', 'beds', 'baths', 'borough', 'hood', \
                           'elevator', 'doorman', 'laundry', 'furnished', 'sqft', 'page'])
```

```python
#Make sure boroughs are all Brooklyn, Bronx, Manhattan, Queens or Staten Island
set(df['borough'])
```

```python
#Examine to see what needs to be removed and what fixed.
group = df.groupby(['borough'])
for key, values in group:
    print(key) #this indicates the grouping
    print('-'*70)
    print(values) #this is a dataframe for that 
    print('\n')
```

```python
#Clean up boroughs
manhattan = ['14th to 40th Street', 'Midtown', 'Midtown Manhattan', 'Murray Hill', 'meat packing', \
             'Upper Manhattan', 'Financial District', 'Greenwich Village']
remove = ['Greenburgh', 'Hempstead', 'Jersey City', 'Kirkland', 'Newark', 'North Hempstead', \
         'Placer County', 'Rye', 'Weehawken', 'Yakima County', 'Yonkers']
df['borough'] = df['borough'].apply(lambda x: 'Manhattan' if x in manhattan else x)
df['borough'] = df['borough'].apply(lambda x: 'Bronx' if x in ['Town'] else x)
df['borough'] = df['borough'].apply(lambda x: 'Brooklyn' if x in ['Tabernacle'] else x)
df = df[~df['borough'].isin(remove)]
set(df['borough'])
```

```python
#df['hood'] = df['hood'].apply(lambda x: 'Crown Heights' if x in ['Tabernacle'] else x )
df.loc[df['hood'] == 'Tabernacle', 'hood'] = 'Crown Heights'
df.loc[df['hood'] == 'Beverley Square West', 'hood'] = 'Flatbush'
```

```python
df.to_csv('rent_info_cleaned.csv')
```
